# Names

```python
FF = [1,1,3,2,2,0,3,3]
FM = [5,5,3,0,0,0,1,2]
MF = [1,6,0,0,0,0,0,1]
MM = [8,4,3,7,0,11,7,6]
names = {
    'FF' : [1,1,3,2,2,0,3,3],
    'FM' : [5,5,3,0,0,0,1,2],
    'MF' : [1,6,0,0,0,0,0,1],
    'MM' : [8,4,3,7,0,11,7,6]
}
df = pd.DataFrame(names)
```

```python
x = np.arange(len(FF))
width = 0.2
fig, ax = plt.subplots()

ax.bar(x - 1.5*width, FF, width, label='FF',color='purple')
ax.bar(x - 0.5*width, FM, width, label='FM',color='gold')
ax.bar(x + 0.5*width, MF, width, label='MF',color='indigo')
ax.bar(x + 1.5*width, MM, width, label='MM',color='yellow')
ax.legend()
```

```python
x = np.arange(len(FF))
width = 0.4
fig, ax = plt.subplots()
ax.bar(x - width/2, FF, width, label='FF', color='purple')
ax.bar(x + width/2, FM, width, label='FM', color='gold')
ax.legend()
```
